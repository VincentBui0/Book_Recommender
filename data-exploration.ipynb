```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing values")

plt.show()
```

```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_books"] = 2024 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_books", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method = "spearman")

sns.set_theme(style="white")
plt.figure(figsize = (8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws ={"label": "Spearman correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()
```

```python
book_missing = books[~(books["description"].isna()) &
~(books["num_pages"].isna()) &
~(books["average_rating"].isna()) &
~(books["published_year"].isna())
]
```

```python
book_missing
```

```python
book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
book_missing
```

```python
book_missing.loc[:, "words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
books_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
```

```python
books_missing_25_words
```

```python
books_missing_25_words.loc[:, "title_and_subtitle"] = (
    np.where(books_missing_25_words["subtitle"].isna(), books_missing_25_words["title"], books_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
books_missing_25_words
```

```python
books_missing_25_words.loc[:, "tagged_description"] = books_missing_25_words[["isbn13", "description"]].astype(str).agg(": ".join, axis=1)
```

```python
books_missing_25_words
```

```python
(
    books_missing_25_words.drop(["subtitle", "missing_description", "age_of_books", "words_in_description"], axis = 1).to_csv("books_cleaned.csv", index = False)
)
```
